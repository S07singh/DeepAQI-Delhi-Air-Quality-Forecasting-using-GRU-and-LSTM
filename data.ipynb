```python
from datasets import load_dataset
import pandas as pd
```

```python
dataset = load_dataset("abhinavsarkar/delhi_air_quality_feature_store_processed.csv")
df = dataset["train"].to_pandas()
```

```python
# Convert timestamp and sort chronologically
df["event_timestamp"] = pd.to_datetime(df["event_timestamp"])
df = df.sort_values("event_timestamp").reset_index(drop=True)
```

```python
# Take the latest 5000 rows
df_latest = df.tail(5000).reset_index(drop=True)
print(f"Shape: {df_latest.shape}")
print(f"Date range: {df_latest['event_timestamp'].min()} to {df_latest['event_timestamp'].max()}")
print(f"Stations: {df_latest['city'].nunique()}")
```

```python
# Save to data folder
df_latest.to_csv("data/latest_5000_rows.csv", index=False)
df_latest.to_parquet("data/latest_5000_rows.parquet", index=False)
print("Saved to data/latest_5000_rows.csv and data/latest_500_rows.parquet")
```
